```julia
#DGP 
function dgp(n)
     
    u= rand(Uniform(0,1),n)
    x= rand(Chi(1),n)
    
    #Beta for each percentile 
    β= u.-0.5
    
    #Generate Y 
    y= x.*β
    
    return y, x, u
    
end
```

### (b)

```julia
function quantEst(n, y, x, τ)
    
    #Number of vars?
    J=2*n
    
    #Number of equations
    K=n
    
    #Declaring the model 
    M=Model(HiGHS.Optimizer)
    set_silent(M)
    
    #Declare the optimization variables
    @variable(M, z[1: J] >= 0.0)
    @variable(M, -100 <= b <= 100)
    
    #Define the objective function
    c = [τ .* ones(n,1); (1-τ) .* ones(n,1)]
    @objective(M, Min, sum(c[j] * z[j] for j in 1:J))

    #Define constraint
    aux= Matrix(I,n,n) #identity matrix
    E= [aux (-1).*aux]
    e= y
    @constraint(M, constraint[k in 1:K], sum(E[k,j]*z[j] for j in 1:J) + x[k]*b == y[k])
    
    # Solve the optimization problem
    optimize!(M)
    b_minimizer = value(b)
    
end
```

### (c)

```julia
Random.seed!(1123);

n= 100 
Y, X, U =dgp(n);
```

### (d)

```julia
#Estimate for different taus
T=zeros(3)
i=1
for q in [0.25 0.50 0.75]
    b = quantEst(n, Y, X, q)
    T[i]=b
    println("The estimator of β(τ) with τ=", q, " is ", b)
    
    i=i+1
end
```

### (e)

```julia
#PLot of lineal fitting for different quantiles conditional on X
plot(X, X*T[1], label="τ=0.25")
plot!(X, X*T[2],  label="τ=0.50")
plot!(X, X*T[3], label="τ=0.75")
```
